```python
df.drop(["ID", "Z_Revenue", "Z_CostContact"], axis = "columns", inplace = True)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_scaled = pd.DataFrame(scaler.fit_transform(df), columns = df.columns)
```

```python
df_scaled.describe().T
```

```python
x = df_scaled.drop("Response", axis = "columns").copy().values
y = df_scaled["Response"].copy().values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, stratify = y, random_state = 17)
```

```python
import torch
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
```

```python
x_train, x_test = torch.FloatTensor(x_train), torch.FloatTensor(x_test)
y_train, y_test = torch.LongTensor(y_train), torch.LongTensor(y_test)
```

```python
train_data = TensorDataset(x_train, y_train)
test_data = TensorDataset(x_test, y_test)
```

```python
train_loader = DataLoader(train_data, batch_size = 10, shuffle = True)
test_loader = DataLoader(test_data, batch_size = 10, shuffle = True)
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
Model = nn.Sequential(
    nn.Linear(38, 256),
    nn.ReLU(),
    nn.Dropout(0.3),
    nn.Linear(256, 128),
    nn.ReLU(),
    nn.Dropout(0.5),
    nn.Linear(128, 128),
    nn.ReLU(),
    nn.Dropout(0.5),
    nn.Linear(128, 84),
    nn.ReLU(),
    nn.Dropout(0.5),
    nn.Linear(84, 2),
    nn.Softmax(dim = 1)
)
```

```python
Model.to(device)
```

```python
loss_function = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(Model.parameters(), lr = 0.001)
```

```python
epochs = 100
training_losses = []
testing_losses = []
training_accuracies = []
testing_accuracies = []

for epoch in range(epochs):

  training_correct = 0
  for _, (x_train, y_train) in enumerate(train_loader):
    y_pred = Model(x_train.to(device))
    loss = loss_function(y_pred, y_train.to(device))

    training_correct += (torch.max(y_pred, 1)[1] == y_train.to(device)).sum()

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    training_losses.append(loss.item())
  training_accuracy = training_correct.item() / len(train_data)
  training_accuracies.append(training_accuracy)

  testing_correct = 0
  with torch.no_grad():
    for _, (x_test, y_test) in enumerate(test_loader):
      y_pred = Model(x_test.to(device))
      loss = loss_function(y_pred, y_test.to(device))
      testing_losses.append(loss.item())
      testing_correct += (torch.max(y_pred, 1)[1] == y_test.to(device)).sum()
  testing_accuracy = testing_correct / len(test_data)
  testing_accuracies.append(testing_accuracy)

  if (epoch + 1) % 5 == 0:
    print(f"Epoch : {epoch + 1} | Training Acc : {training_accuracy} | Testing Acc : {testing_accuracy}")

  if (epoch + 1) == epochs:
    print(f"\n\nLowest Training Loss : {np.min(training_losses)} \nLowest Testing Loss : {np.min(testing_losses)}")
    print(f"\n\nHighest Training Loss : {np.max(training_losses)} \nHighest Testing Loss : {np.max(testing_losses)}")
```

```python
y_testing = []
y_predicted = []

with torch.no_grad():
  correct = 0
  for X_test, y_test in test_loader:
    y_pred = Model(X_test.to(device))
    predicted = torch.max(y_pred, 1)[1]
    for i in predicted:
      y_predicted.append(i.to("cpu"))

    for i in y_test:
      y_testing.append(i.to("cpu"))

    correct += (predicted == y_test.to(device)).sum()

# Accuracy
print(f"Accuracy : {correct.item() / len(test_data) * 100}%")
```
